```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Loading the dataset
# Note: Ensure 'US_Accidents_March23.csv' is uploaded to your Colab 'Files' tab
df = pd.read_csv('US_Accidents_March23.csv', nrows=100000)

# 2. Cleaning & Processing
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['Hour'] = df['Start_Time'].dt.hour
df['Weekday'] = df['Start_Time'].dt.day_name()

# 3. Statistical Insight: Accidents by Hour
plt.figure(figsize=(10, 5))
sns.histplot(df['Hour'], bins=24, color='royalblue', kde=True)
plt.title('Accident Frequency by Time of Day')
plt.show()
```

```python
# Step 8: Interactive Dashboard
# This allows stakeholders to hover and explore specific regions
fig = px.scatter_mapbox(df.sample(2000),
                        lat="Start_Lat",
                        lon="Start_Lng",
                        color="Severity",
                        size="Severity",
                        color_continuous_scale=px.colors.sequential.OrRd,
                        zoom=3,
                        mapbox_style="carto-positron",
                        title="Interactive DOT Accident Map")
fig.show()
```
